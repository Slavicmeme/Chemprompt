```python
from tqdm import tqdm
from Chemprompt.embeddings.finger_print import FingerprintGenerator
from Chemprompt.data.data_loader import DataLoader
from Chemprompt.models.sklearn_model import ScikitLearnModel
```

```python
dataloader = DataLoader()
```

```python
data_dict = {
    "Caco2_Wang": "regression",
    "FreeSolv": "regression",
    "ESOL": "regression",
    "Lipo": "regression",
    "HPPB": "regression",
}

finger_print_type = ["Morgan", "MACCS", "RDKit", "AtomPair", "TopologicalTorsion", "Pattern"]
```

```python
for dataset, dtype in tqdm(data_dict.items(), desc="Processing Datasets"):
    smiles_list, y = dataloader.load_dataset(dataset)

    for finger_print_name in tqdm(finger_print_type, desc=f"Processing {dataset}", leave=False):
        generator = FingerprintGenerator(fp_type=finger_print_name)
        x = generator.transform(smiles_list)
        sk_model = ScikitLearnModel(model_type=dtype, save_dir=f"./result/finger_print/{dtype}/{dataset}/{finger_print_name}/")
        sk_model.fit_and_evaluate_fold(smiles_list, x, y)
```
