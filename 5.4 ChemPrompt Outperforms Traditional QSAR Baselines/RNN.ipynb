```python
from Chemprompt.models.nlp_model import ModelTrainer, LSTMClassifier
from Chemprompt.data.data_loader import DataLoader
```

```python
DATASETS = ["FreeSolv"]
```

```python
loader = DataLoader()
```

```python
for dataset_name in DATASETS:
    print(f"\nProcessing dataset: {dataset_name}")

    # data load
    x, y = loader.load_dataset(dataset_name)
    experiments = {
        "raw_smiles": x
    }

    # LSTM & RNN train and evaluation
    for model_type in ["lstm", "rnn"]:
        for exp_name, smiles_data in experiments.items():
            exp_full_name = f"{dataset_name}_fix_{exp_name}_Regression_{model_type}"
            print(f"\nRunning experiment: {exp_full_name} (Model: {model_type}, Task: Regression)")
            
            trainer = ModelTrainer(
                model_type=model_type,
                task_type="Regression",
                x=smiles_data,
                y=y,
                data_name=exp_full_name,
                num_epochs=100,
                lr=1e-4,
                device = "cuda:1"
            )
            trainer.train_and_evaluate()
```
